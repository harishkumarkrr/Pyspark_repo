```python
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("pysparkLearn").getOrCreate()
```

```python
df = spark.read.csv("US-Gun-Violence.csv", header=True)
```

```python
df.show(5)
```

```python
df.filter(df.injured.isNotNull()).show(5)
```

```python
df.createOrReplaceTempView('DATA')
```

```python
spark.sql("select state,city_or_county,killed from DATA where killed>10 ").show()
```
