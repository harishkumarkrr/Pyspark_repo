```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("weather_api").getOrCreate()
```

```python
df = spark.read.option("multiLine","true").json('weather.json')
```

```python
df.show()
```

```python
import json
with open('weather.json') as f:
    output = json.load(f)
    
jout = output['dataseries']
```
