```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col


spark = SparkSession.builder.appName("pysparkLearn").getOrCreate()
```

```python
df = spark.read.csv("US-Gun-Violence.csv", header=True)
```

```python
df.show(5)
```

```python
df.groupby('state','killed').count().sort(col('count').desc()).filter(df.killed>10).show()
```
